```python
import gradio as gr
import pandas as pd

def upload_csv(file):
    df = pd.read_csv(file.name)
    return df

with gr.Blocks(title="CSV Upload in Gradio") as demo:
    gr.Markdown("## ðŸ“‚ Upload Hostel Fee CSV File")

    file_input = gr.File(label="hostel_fee_payment_dataset.csv", file_types=[".csv"])
    output_df = gr.Dataframe(label="CSV Preview")

    file_input.change(
        upload_csv,
        inputs=file_input,
        outputs=output_df
    )

demo.launch(share=True)
```
