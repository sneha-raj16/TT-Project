```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df3=pd.DataFrame(data3)
df4=pd.DataFrame(data4)

df=pd.concat([df1,df2,df3,df4],ignore_index=True)

df["Due_Date"] = pd.to_datetime(df["Due_Date"])
df["Payment_Date"] = pd.to_datetime(df["Payment_Date"])

df["Delay_Days"] = (df["Payment_Date"] - df["Due_Date"]).dt.days
df["Delay_Days"] = df["Delay_Days"].fillna(-1)

def compilance_status(days):
  if days == -1:
    return "Not Paid"
  elif days == 0:
    return "On Time"
  elif days < 0:
    return "Early"
  else:
    return "Late"

df["Payment_Status"] = df["Delay_Days"].apply(compilance_status)

print(df)
```

```python
print("Average Delay:", df[df["Delay_Days"] >=0]["Delay_Days"].mean())
print("Maximum Delay:", df["Delay_Days"].max())
print("Minimum Delay:", df["Delay_Days"].min())

print("\nPayment Status Count:")
print(df["Payment_Status"].value_counts())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Payment_Status", data=df)
plt.title("Payment Status Distribution of Students")
plt.xlabel("Payment Status")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="Student_Name", y="Fee_Amount", data=df)
plt.title("Student-wise Hostel Fee Amount")
plt.xlabel("Student Name")
plt.ylabel("Fee Amount (â‚¹)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    x="Student_Name",
    y="Delay_Days",
    hue="Payment_Status",
    data=df
)
plt.title("Student-wise Payment Delay with Status")
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = df["Payment_Status"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    status_counts,
    labels=status_counts.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Hostel Fee Payment Status Distribution")
plt.show()
```
